```python
pd.options.mode.chained_assignment = None
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (15,5)
def draw_plot(config, data):
    x = []
    y = []
    for i,j in data.items():
        x.append(str(i.left))
        y.append(j)
    plt.scatter(x,y)
    plt.show()
    
data_to_plot = data.groupby(["buff size", "Producer/Consumer", "prod cons config",  "is randomized","fairness",])
def show_plots():
    for group,plot_data in data_to_plot:
        print(f"Buffer size is {group[0]}, {group[1]} -> {group[2]}, fairness: {group[4]}, randomized: {group[3]}")
        min_size = plot_data['size'].min()
        max_size = plot_data['size'].max()
        interval = 200
        if(max_size > 5000):
            interval = 1500
        plot_data['bin'] = pd.cut(plot_data["size"], [x for x in range(0, max_size+100, interval)])
        a = plot_data.groupby('bin')["counter"].sum()
        print()
        yield draw_plot(group, a)
```

```python
iterator = show_plots()
next(iterator)
next(iterator)
```

Oba wykresy wyglądają niemal identycznie, jednak dla wersji sprawiedliwej konsumncja jest niższa dla analogicznych wartości z przypadku normalnego buffora, dodatkowo widzimy że wartości mniejszych jest więcej zgodnie z tym aby prawdopodobnieństwo wylosowania mniejszej wartości było większe.
